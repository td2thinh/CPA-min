```python
from matplotlib import pyplot as plt
from naive import naiveAlgorithm
from minidisk import minidisk
from readPoints import readPointsInstance, readPointsFromFile
import sys
from time import perf_counter_ns, perf_counter

sys.setrecursionlimit(10000)
```

## Measuring execution time of each instance over 100 instances

```python
exec_time_naive = []
circle_naive = []
for i in range(2, 3):
    points = readPointsFromFile(f"input/test-{i}.points")
    t1_start = perf_counter()
    circle = naiveAlgorithm(points)
    t1_stop = perf_counter()
    circle_naive.append(circle)
    exec_time_naive.append(t1_stop - t1_start)
```

```python
exec_time_minidisk = []
circle_minidisk = []
for i in range(2, 3):
    points = readPointsFromFile(f"input/test-{i}.points")
    t2_start = perf_counter()
    cirlce = minidisk(points)
    t2_stop = perf_counter()
    circle_minidisk.append(circle)
    exec_time_minidisk.append(t2_stop - t2_start)
```

## Execution time of each instance

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.tight_layout(pad=5.0)
ax1.set_title("Naive algorithm")
ax1.plot(range(2, 3), exec_time_naive, label="Execution time")
ax1.set_xlabel("Instance")
ax1.set_ylabel("Execution time")
ax2.set_title("Welzl's algorithm")
ax2.plot(range(2, 3), exec_time_minidisk, label="Execution time", color="orange")
ax2.set_xlabel("Instance")
ax2.set_ylabel("Execution time")
```

## Average time of execution

```python
avg_time_naive = sum(exec_time_naive) / len(exec_time_naive)
avg_time_minidisk = sum(exec_time_minidisk) / len(exec_time_minidisk)
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.tight_layout(pad=5.0)
ax1.set_title("Average execution time")
ax1.bar(
    ["Naive", "Welzl"],
    [avg_time_naive, avg_time_minidisk],
    label="Average execution time",
    color=["blue", "orange"],
)
ax1.set_ylabel("Average execution time")
ax2.set_title("Welzl's algorithm")
ax2.bar("Welzl", avg_time_minidisk, label="Average execution time", color="orange")
ax2.set_ylabel("Average execution time")
```
